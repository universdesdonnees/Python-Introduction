## **Exercice 1 : L'Aventure des Animaux**
Objectif : Explorer les fonctions de base de Pandas (création, sélection, filtrage).
Données : Créez un DataFrame animaux contenant les colonnes Nom, Espèce, Age et Poids.

3. Ajoutez une nouvelle colonne Longévité où vous estimez la durée de vie maximale de chaque animal.

## **Exercice 3 : Le Mystère des Fruits**
Objectif : Pratiquer le slicing, le filtrage, et la manipulation de chaînes.
Données : Créez un DataFrame fruits avec les colonnes Nom, Couleur, Prix.

2. Sélectionnez tous les fruits de couleur rouge.
3. Augmentez le prix de tous les fruits de 20%.
4. Ajoutez une colonne Nom_Majuscule qui contient les noms des fruits en majuscules.


```python
# Création du DataFrame
data = {
    "Nom": ["Pomme", "Banane", "Cerise", "Orange"],
    "Couleur": ["Rouge", "Jaune", "Rouge", "Orange"],
    "Prix": [1.00, 0.50, 1.50, 0.80]
}
fruits = pd.DataFrame(data)

# Sélection des fruits de couleur rouge
fruits_rouges = fruits[fruits["Couleur"] == "Rouge"]

# Augmentation du prix de 20%
fruits["Prix"] *= 1.20

# Ajout de la colonne Nom_Majuscule
fruits["Nom_Majuscule"] = fruits["Nom"].str.upper()
fruits
```

## **Exercice 4 : Le Festival de Musique**
Objectif : Fusionner des DataFrames et gérer des données manquantes.
Données : Créez deux DataFrames, artistes avec les colonnes Nom, Genre et Année, et concerts 
avec les colonnes Artiste, Lieu, Date.
Tâches :
1. Ajoutez des artistes et leurs informations dans artistes.
2. Ajoutez des concerts dans concerts, en faisant correspondre certains artistes.
3. Fusionnez les deux DataFrames pour avoir une vue complète des concerts avec les détails des artistes.
4. Remplacez toutes les données manquantes par des valeurs par défaut (par exemple, "Inconnu" pour le lieu).



```python
# Création du DataFrame 'artistes'
data_artistes = {
    "Nom": ["The Beatles", "Lady Gaga", "Queen", "Ed Sheeran"],
    "Genre": ["Rock", "Pop", "Rock", "Pop"],
    "Année": [1960, 2008, 1970, 2011]
}
artistes = pd.DataFrame(data_artistes)

# Création du DataFrame 'concerts'
data_concerts = {
    "Artiste": ["The Beatles", "Queen", "Ed Sheeran", "Adele"],
    "Lieu": ["Wembley", "Madison Square Garden", None, "O2 Arena"],
    "Date": ["2024-06-30", "2024-07-02", "2024-07-04", "2024-07-05"]
}
concerts = pd.DataFrame(data_concerts)

# Fusion des DataFrames
fusion = pd.merge(concerts, artistes, left_on="Artiste", right_on="Nom", how="left")

# Remplacement des valeurs manquantes None -> 'Inconnu'
fusion.fillna("Inconnu", inplace=True)
fusion
```
