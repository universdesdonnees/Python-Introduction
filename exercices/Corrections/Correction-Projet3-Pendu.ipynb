```python
def display_hangman(tries):
    stages = [  # Final state: head, torso, both arms, and both legs
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                   -
                """,
                # Head, torso, both arms, and one leg
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / 
                   -
                """,
                # Head, torso, and both arms
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |      
                   -
                """,
                # Head, torso, and one arm
                """
                   --------
                   |      |
                   |      O
                   |     \\|
                   |      |
                   |     
                   -
                """,
                # Head and torso
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                """,
                # Head
                """
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                """,
                # Initial empty state
                """
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                """
    ]
    return stages[tries]
```

```python
import random

def hangman():
    mots = ["chat", "souris", "table", "elephant"]
    mot_tire = random.choice(mots)
    mot_cache = ["_"] * len(mot_tire)
    tries = 6
    deviner = False

    print("Jeu du Pendu!")
    print(display_hangman(tries))
    print(" ".join(mot_cache))

    while not deviner and tries > 0:
        lettre_devinee = input("Devinez une lettre: ").lower()

        if lettre_devinee in mot_tire and lettre_devinee not in mot_cache:
            print("Bonne lettre!")
            for index, letter in enumerate(mot_tire):
                if letter == lettre_devinee:
                    mot_cache[index] = lettre_devinee

            if "_" not in mot_cache:
                deviner = True
        else:
            tries -= 1
            print("Faux! Il vous reste {} essais.".format(tries))
            print(display_hangman(tries))

        print(" ".join(mot_cache))

    if deviner:
        print("Félicitations! Vous avez deviné le mot: {}".format(mot_tire))
    else:
        print("Désolé, vous avez perdu. Le mot était: {}".format(mot_tire))

hangman()
```
