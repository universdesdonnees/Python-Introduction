## Exercice 1 : Puissance 4

- CrÃ©ez une fonction qui crÃ©e une grille de 7 par 6 contenant uniquement des zÃ©ros.
- CrÃ©ez une fonction qui ajoute une piÃ¨ce selon la ligne et la colonne qu'on lui fournit
- VÃ©rifiez que l'emplacement oÃ¹ vous ajoutez une piÃ¨ce est valide (ne contient pas de piÃ¨ce / est dans la grille)
- VÃ©rifiez aussi Ã  l'aide d'une fonction la prochaine ligne disponible par rapport Ã  la colonne qu'on lui fournit.

BONUS 2: Remplacer les piÃ¨ces placÃ©es dans la grille par des emojis ou des carrÃ©s rouges et jaunes comme dans le jeu original

```python
import numpy as np

def initiation_grille():
    return np.zeros((7, 6), dtype=int)

def jouer_coup(grille, colonne, joueur):
    # VÃ©rifie si la colonne est dans la plage valide
    if 0 <= colonne < 6:  # VÃ©rifie si la colonne est valide (0 Ã  5)
        for l in range(6):  # Parcourt la colonne de haut en bas
            if grille[l][colonne] == 0:
                grille[l][colonne] = joueur
                return grille
    else:
        print("Colonne invalide. Veuillez sÃ©lectionner une colonne entre 0 et 5.")
        return None  # Retourne None pour indiquer une erreur

def verifier_gagnant(grille):
    # VÃ©rifier les lignes pour un alignement de 4 jetons
    for i in range(7):
        for j in range(3):
            if grille[i, j] == grille[i, j+1] == grille[i, j+2] == grille[i, j+3] != 0:
                return grille[i, j]

    # VÃ©rifier les colonnes pour un alignement de 4 jetons
    for j in range(6):
        for i in range(4):
            if grille[i, j] == grille[i+1, j] == grille[i+2, j] == grille[i+3, j] != 0:
                return grille[i, j]

    # VÃ©rifier les diagonales (haut-gauche vers bas-droite)
    for i in range(4):
        for j in range(3):
            if grille[i, j] == grille[i+1, j+1] == grille[i+2, j+2] == grille[i+3, j+3] != 0:
                return grille[i, j]

    # VÃ©rifier les diagonales (bas-gauche vers haut-droite)
    for i in range(3, 7):
        for j in range(3):
            if grille[i, j] == grille[i-1, j+1] == grille[i-2, j+2] == grille[i-3, j+3] != 0:
                return grille[i, j]

    return 0  # Pas de gagnant trouvÃ©
   
def afficher_grille(grille):
    # DÃ©finir les reprÃ©sentations pour chaque joueur
    representation_joueur1 = "ðŸ”´"  # Emoji point rouge
    representation_joueur2 = "ðŸŸ¡"  # Emoji point jaune
    representation_vide = "âšª"     # Emoji point blanc pour les cellules vides

    # Parcourir la grille et afficher l'emoji correspondant
    grille = np.flip(grille, 0)
    for ligne in grille:
        for cellule in ligne:
            if cellule == 1:
                print(representation_joueur1, end=" ")
            elif cellule == 2:
                print(representation_joueur2, end=" ")
            else:
                print(representation_vide, end=" ")
        print()
    print( "\n")  # saut de ligne

def puissance4():
    grille = initiation_grille()
    joueur_actuel = 1

    while True:
        afficher_grille(grille)
        colonne = int(input(f"Joueur {joueur_actuel}, choisissez une colonne (0-5): "))
        grille = jouer_coup(grille, colonne, joueur_actuel)

        if grille is None:
            continue

        if verifier_gagnant(grille):
            afficher_grille(grille)
            print(f"Le joueur {joueur_actuel} a gagnÃ© !")
            break

        # Bascule avec un ternaire
        joueur_actuel = 1 if joueur_actuel == 2 else 2  # Change de joueur

puissance4()
```

## Exercice 2 : Le train

Vous Ãªtes ligoteÌ sur les rails en gare de Toulouse. EÌcrire un programme qui affiche un tableau vous permettant de connaiÌ‚tre lâ€™heure aÌ€ laquelle vous serez deÌchiqueteÌ par le train parti de la gare Montparnasse aÌ€ 9h (il y a 674 km entre la gare Montparnasse et Toulouse).

Le tableau preÌdira les diffeÌrentes heures possibles pour toutes les vitesses de 100 km/h aÌ€ 300 km/h, par pas de 20 km/h, les reÌsultats eÌtant arrondis aÌ€ la minute infeÌrieure.

```python
# Initialisation
depart = 9  # Heure de dÃ©part
d = 674  # Distance en km

# En-tÃªte pour l'affichage
print("Vitesse (km/h) | Heure du dÃ©chictage")

# Boucle sur la plage de vitesses
for v in range(100, 301, 20):
    # Calculer le temps pris en heures
    temps_pris = d / v
    
    # Calculer l'heure d'arrivÃ©e
    arrivee = depart + temps_pris
    heure_arrivee = int(arrivee)
    min_arrivee = int((arrivee - heure_arrivee) * 60) # fraction d'heure convertie en minutes

    # Affichage du rÃ©sultat
    print(f"{v} km/h       |       {heure_arrivee}h{min_arrivee:02}min")
```
