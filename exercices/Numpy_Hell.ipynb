## Exercice 1 : Puissance 4

- Créez une fonction qui crée une grille de 7 par 6 contenant uniquement des zéros.
- Créez une fonction qui ajoute une pièce selon la ligne et la colonne qu'on lui fournit
- Vérifiez que l'emplacement où vous ajoutez une pièce est valide (ne contient pas de pièce / est dans la grille)
- Vérifiez aussi à l'aide d'une fonction la prochaine ligne disponible par rapport à la colonne qu'on lui fournit.
- Mettre en place une fonction qui vérifie si 4 pièces sont allignées (Toutes les directions)
BONUS : Optimiser votre code (Noms des joueurs, Gestion des exceptions)
BONUS 2: Remplacer les pièces placées dans la grille par des emojis ou des carrés rouges et jaunes comme dans le jeu original

```python
import numpy as np

def initiation_grille():
    return np.zeros((7, 6), dtype=int)

def jouer_coup(grille, colonne, joueur):
    # Vérifie si la colonne est dans la plage valide
    if 0 <= colonne < 6:  # Vérifie si la colonne est valide (0 à 5)
        for l in range(6):  # Parcourt la colonne de haut en bas
            if grille[l][colonne] == 0:
                grille[l][colonne] = joueur
                return grille
    else:
        print("Colonne invalide. Veuillez sélectionner une colonne entre 0 et 5.")
        return None  # Retourne None pour indiquer une erreur

def verifier_gagnant(grille):
    # Vérifier les lignes pour un alignement de 4 jetons
    for i in range(7):
        for j in range(3):
            if grille[i, j] == grille[i, j+1] == grille[i, j+2] == grille[i, j+3] != 0:
                return grille[i, j]

    # Vérifier les colonnes pour un alignement de 4 jetons
    for j in range(6):
        for i in range(4):
            if grille[i, j] == grille[i+1, j] == grille[i+2, j] == grille[i+3, j] != 0:
                return grille[i, j]

    # Vérifier les diagonales (haut-gauche vers bas-droite)
    for i in range(4):
        for j in range(3):
            if grille[i, j] == grille[i+1, j+1] == grille[i+2, j+2] == grille[i+3, j+3] != 0:
                return grille[i, j]

    # Vérifier les diagonales (bas-gauche vers haut-droite)
    for i in range(3, 7):
        for j in range(3):
            if grille[i, j] == grille[i-1, j+1] == grille[i-2, j+2] == grille[i-3, j+3] != 0:
                return grille[i, j]

    return 0  # Pas de gagnant trouvé
   
def afficher_grille(grille):
    # Définir les représentations pour chaque joueur
    representation_joueur1 = "🔴"  # Emoji point rouge
    representation_joueur2 = "🟡"  # Emoji point jaune
    representation_vide = "⚪"     # Emoji point blanc pour les cellules vides

    # Parcourir la grille et afficher l'emoji correspondant
    grille = np.flip(grille, 0)
    for ligne in grille:
        for cellule in ligne:
            if cellule == 1:
                print(representation_joueur1, end=" ")
            elif cellule == 2:
                print(representation_joueur2, end=" ")
            else:
                print(representation_vide, end=" ")
        print()
    print( "\n")  # saut de ligne

def puissance4():
    grille = initiation_grille()
    joueur_actuel = 1

    while True:
        afficher_grille(grille)
        colonne = int(input(f"Joueur {joueur_actuel}, choisissez une colonne (0-5): "))
        grille = jouer_coup(grille, colonne, joueur_actuel)

        if grille is None:
            continue

        if verifier_gagnant(grille):
            afficher_grille(grille)
            print(f"Le joueur {joueur_actuel} a gagné !")
            break

        # Bascule avec un ternaire
        joueur_actuel = 1 if joueur_actuel == 2 else 2  # Change de joueur

puissance4()
```

## Exercice 2 : Le train

Vous êtes ligoté sur les rails en gare de Toulouse. Écrire un programme qui affiche un tableau vous permettant de connaître l’heure à laquelle vous serez déchiqueté par le train parti de la gare Montparnasse à 9h (il y a 674 km entre la gare Montparnasse et Toulouse).

Le tableau prédira les différentes heures possibles pour toutes les vitesses de 100 km/h à 300 km/h, par pas de 20 km/h, les résultats étant arrondis à la minute inférieure.

```python
# Initialisation
depart = 9  # Heure de départ
d = 674  # Distance en km

# En-tête pour l'affichage
print("Vitesse (km/h) | Heure du déchictage")

# Boucle sur la plage de vitesses
for v in range(100, 301, 20):
    # Calculer le temps pris en heures
    temps_pris = d / v
    
    # Calculer l'heure d'arrivée
    arrivee = depart + temps_pris
    heure_arrivee = int(arrivee)
    min_arrivee = int((arrivee - heure_arrivee) * 60) # fraction d'heure convertie en minutes

    # Affichage du résultat
    print(f"{v} km/h       |       {heure_arrivee}h{min_arrivee:02}min")
```

### Exercice 3 : Binaire
Créer un programme pour convertir un nombre binaire en un nombre décimal et inversement.

```python
def convertion_binaire(nombre_decimal):
    return np.binary_repr(nombre_decimal)

def convertion_decimal(nombre_binaire):
    return int(nombre_binaire, 2)

print(f"563 en binaire est : {convertion_binaire(563)}"),  
print(f"1000110011 converti en décimal est : {convertion_decimal('1000110011')}")
```
